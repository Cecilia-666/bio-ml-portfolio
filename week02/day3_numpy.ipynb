```python
import numpy as np

# ---NumPy arrays vs Python lists---
expression = np.array([12.5, 8.3, 45.2, 3.1, 22.8, 16.4, 9.7, 33.2])

# Basic statistics — done in one line
print(f"Mean value: {np.mean(expression):.2f}")
print(f"Standard deviation: {np.std(expression):.2f}")
print(f"Median value: {np.median(expression):.2f}")
print(f"Minimum value: {np.min(expression):.2f}")
print(f"Maximum value: {np.max(expression):.2f}")

#---Vectorized operations---
log2_expression = np.log2(expression + 1)
print(f"\nAfter Log2 transformed: {log2_expression.round(2)}")

z_scores = (expression - np.mean(expression)) / np.std(expression)
print(f"Z-score: {z_scores.round(2)}")
```

```python
# ── 2D array: Gene expression matrix ────────────────────────────
# Rows = genes, columns = samples — this is the basic structure of transcriptomic data

expression_matrix = np.array([
    [12.5, 13.1, 11.8],   # 基因1，3个重复
    [45.2, 43.8, 46.1],   # 基因2
    [8.3,   9.1,  7.9],   # 基因3
    [22.1, 21.5, 23.0],   # 基因4
])
print(f"The shape of the matrix (gene number, sample number): {expression_matrix.shape}")

# Calculate the mean of each gene, axis=1
gene_means = np.mean(expression_matrix, axis=1)
print(f"The mean of each gene is {gene_means.round(2)}")

# Calculate the mean of each column, axis=0
sample_means = np.mean(expression_matrix, axis=0)
print(f"The mean of each sample is {sample_means.round(2)}")

# Select each row/column
print(f"\nSelect the expression of gene2 {expression_matrix[1, :]}")
print(f"\nSelect the expression of sample1 {expression_matrix[:, 0]}")
```

```python
# ---The eighth chapter -  List comprehension---
sequences = ["ATGCGC", "ATATAT", "GCGCGC", "ATCGATCG"]

# Traditional for-loop approach
gc_list_old = []
for seq in sequences:
    gc = round((seq.count("G") + seq.count("C")) / len(seq) * 100, 2)
    gc_list_old.append(gc)

# List comprehension
gc_list_new = [round((seq.count("G") + seq.count("C")) / len(seq) * 100, 2) for seq in sequences]

print(f"Traditional for-loop: {gc_list_old}")
print(f"List comprehension: {gc_list_new}")

# Filter: retain sequences with GC content > 50%
high_gc = [seq for seq in sequences
           if (seq.count("G") + seq.count("C")) / len(seq) > 0.5]
print(f"High GC sequences: {high_gc}")
```

```python
# Chapter 9: Exception Handling
def safe_gc(sequence):
    """Calculate the safe GC content of a sequence, handle various exceptions"""
    try:
        gc = (sequence.count("G") + sequence.count("C")) / len(sequence) * 100
        return round(gc, 2)
    except ZeroDivisionError:
        print("Warning: Sequence Length is 0")
        return 0
    except AttributeError:
        print("Warning: Invalid sequence input")
        return None

print(safe_gc("ATGC"))
print(safe_gc(""))
print(safe_gc(None))
```
