```python
from email.header import UTF8

import pandas

# ---Rabbits and Recurrence Relations---
def rabbit_count(n, k):
    """
    Calculate the total number of rabbit pairs after n months.
    This function uses recursion to solve the Fibonacci rabbit problem.
    The recurrence relation is: F(n) = F(n-1) + k * F(n-2)

    Parameters:
        n (int): Number of months (1 <= n <= 40)
        k (int): Number of offspring pairs produced by each
                 mature rabbit pair per month (1 <= k <= 5)

    Returns:
        int: Total number of rabbit pairs in month n
    """
    if n <= 2:
        return 1
    else:
        return rabbit_count(n-1, k) + k * rabbit_count(n-2, k)

rabbit_count(33, 5)
```

```python
# ---Computing GC Content---
from Bio import SeqIO
import pandas as pd

# ---Read fasta document---
records = list(SeqIO.parse("rosalind_gc.fasta", "fasta-pearson"))

# ---Calculate GC content with pandas dataframe---
# Transform the sequences into list, and then dataframe
data = []
for record in records:
    gc = (record.seq.count('G') + record.seq.count('C')) / len(record.seq) * 100
    data.append({
        "id": record.id,
        "sequence": record.seq,
        "length": len(record.seq),
        "gc_content": round(gc, 6)
    })

df = pd.DataFrame(data)
print("The GC content of all sequences: ")
display(df[['id', 'length', 'gc_content']])

# ---Find the highest GC content of all sequences---
# Find the row index of the maximum value in the gc_content column
best_idx = df["gc_content"].idxmax()
best_row = df.loc[best_idx]

print(f"\n The answer is:")
print(f"{best_row['id']}")
print(f"{best_row['gc_content']:.6f}")

```

```python
# ---Counting Point Mutations---
# ---HAMM: Calculate the Hamming distance between two sequences---
def hamming_dist(c1, c2):
    """
    Calculate the Hamming distance between two DNA sequences.

    The Hamming distance is the number of positions at which
    the corresponding characters are different.
    Biological meaning: number of point mutations between two sequences.

    Parameters:
        s1 (str): First DNA sequence
        s2 (str): Second DNA sequence (same length as s1)

    Returns:
        int: Number of positions where s1 and s2 differ
    """
    return sum(c1 != c2 for c1, c2 in zip(s1, s2))

with open("rosalind_hamm.txt", "r") as f:
    lines = f.read().strip().split("\n")
    s1 = lines[0]
    s2 = lines[1]

print(f"The length of the first DNA sequence: {len(s1)}")
print(f"The length of the second DNA sequence: {len(s2)}")
print(f"\nThe hamming distance is: {hamming_dist(s1, s2)}")





```
