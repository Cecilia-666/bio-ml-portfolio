```python
from IPython.core.application import programdata

# ── 条件判断进阶 ──────────────────────────────────────
sequences = [
    {"id": "seq1", "gc": 72, "length": 450},
    {"id": "seq2", "gc": 38, "length": 120},
    {"id": "seq3", "gc": 55, "length": 890},
    {"id": "seq4", "gc": 61, "length": 95},
]

for seq in sequences:
    if seq["gc"] > 60 and seq["length"] > 200:
        category = "高GC长序列"
    elif seq["gc"] > 60 and seq["length"] <= 200:
        category = "高GC短序列"
    elif seq["gc"] <= 60 and seq["length"] > 200:
        category = "低GC长序列"
    else:
        category = "低GC短序列"

    print(f"{seq['id']}: GC={seq['gc']}%, 长度={seq['length']}, 分类={category}")
```

```python
# Regular expression
import re

# Basic usage
dna = "ATGCGATCGATGCGATCGATG"

# Find all ATG
matches = re.findall("ATG", dna)
print(f"The time of ATG is {len(matches)} times")
print(matches)

# Find the position of the first "ATG"
match = re.search("ATG", dna)
if match:
    print(f"the position of the first ATG is {match.start()}")
# Special str
protein = "ACDEFGHIKLMNPQRSTVWY"

# Find all charged AA
charged = re.findall("[KRDE]", protein)
print(f"The charged amino acids are {charged}")

#Find Kozak sequence
sequence = "GCGCCACCATGGCTATGCGATGCC"
kozak = re.findall("[AG]CCATGG", sequence)
print(f"The Kozak sequence is {kozak}")

# Find all ATG and their positions
for m in re.finditer("ATG", sequence):
    print(f"The position of ATG is {m.start()} to {m.end()}")
```

```python
# ---Modular programming---
# Function 1: Calculate GC content
def calculate_gc(sequence):
    sequence = sequence.upper()
    if len(sequence) == 0:
        return 0
    else:
        gc = (sequence.count("G") +sequence.count("C")) / len(sequence) * 100
        return round(gc, 2)

# Function 2: Analyze all sequences
def analyze_sequences(sequences):
    """calculate GC content of each sequence, return list"""
    results = []
    for i, seq in enumerate(sequences):
        gc = calculate_gc(seq)
        results.append({"id": f"seq{i+1}", "sequence": seq, "gc_content": gc})
    return results

# Function 3: Format and display the output
def print_results(results):
    """print results"""
    print(f"{"ID":<10}{"GC content":>8}{"length":>8}")
    print("-"*30)
    for r in results:
        print (f"{r["id"]:<10}{r['gc_content']:>6}%{len(r['sequence']):>8}")

# Main program
test_sequences = ["ATGCGCTATCGC", "ATAATATATATAT", "GCGCGCGCGC", "ATCGATCG"]
results = analyze_sequences(test_sequences)
print_results(results)
```

```python
from Bio.Seq import Seq
from Bio import SeqIO

# ---Seq object: A smarter sequence than ordinary strings---
dna = Seq("ATGCGATCGATCGATCG")
print(f"Length of DNA is: {len(dna)}")
print(f"The first 6 bases of DNA are: {dna[:6]}")
print(f"The complement DNA is: {dna.complement()}")
print(f"The reverse complement of DNA is: {dna.reverse_complement()}")
print(f"Transcription (DNA → RNA): {dna.transcribe()}")
print(f"Translate (RNA → Protein): {dna.translate()}")
```
